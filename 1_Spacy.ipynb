### Utilizando o Spacy

```python
documento = nlp("https://link.com Em solenidade com a presença da ministra da Saúde, Nísia Trindade; do prefeito de Goiânia, Rogério Cruz; da reitora da UFG, Angelita Pereira de Lima, e de deputados federais e estaduais, foi anunciada a reestruturação do Hospital Municipal da Mulher e Maternidade Célia Câmara (HMMCC), agora dedicado exclusivamente ao atendimento de especialidades femininas, ampliando e qualificando os serviços voltados à saúde da mulher. O evento ocorreu nesta segunda-feira (1º/7), na sede da Unidade, que é gerida por meio de um convênio tripartite entre Secretaria Municipal de Saúde (SMS), Fundação de Apoio ao Hospital das Clínicas da Universidade Federal de Goiás (Fundahc) e a UFG desde a inauguração, em abril de 2020.")
len(documento), type(documento)
```

##### Matching

```python
from typing import Match
from spacy.matcher import Matcher

padrao = [
    {'ORTH': '('},
    {'SHAPE': 'dd'},
    {'ORTH': ')'},
    {'ORTH': ' ', 'OP': '?'},
    {'ORTH': '-', 'OP': '?'},
    {'IS_DIGIT': True},
    {'ORTH': '-', 'OP': '?'},
    {'IS_DIGIT': True, 'LENGTH': 4, 'OP': '?'},
]

matcher = Matcher(nlp.vocab)
matcher.add('phone', [padrao])
matcher
```

```python
text4 = nlp('Você pode ligar para (62) 987654321 ou para (11) - 12345678')
matches = matcher(text4)
[(id, start, end, text4[start:end]) for (id, start, end) in matches]
```

```python
text5 = nlp('Estamos infectados com micro organismos. MICROORGANISMOS são perigosos. Não enxergamos micro-organismos.')
matcher = Matcher(nlp.vocab)
padroes = [
    [{'LOWER': 'microorganismos'}],
    [{'LOWER': 'micro-organismos'}],
    [
        {'LOWER': 'micro'},
        {'LOWER': 'organismos'},
    ],
]
matcher.add('padrao', padroes)
matches = matcher(text5)
[(id, start, end, text5[start:end]) for (id, start, end) in matches]
```
